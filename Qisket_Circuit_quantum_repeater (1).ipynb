```python
!pip install qiskit qiskit-aer pylatexenc matplotlib -q
!pip install pylatexenc
%matplotlib inline
from qiskit import QuantumCircuit
import matplotlib.pyplot as plt

qc = QuantumCircuit(4, 2)

# Bell pair 1
qc.h(0)
qc.cx(0, 1)

qc.barrier()

# Bell pair 2
qc.h(2)
qc.cx(2, 3)

qc.barrier()

# Entanglement swapping
qc.cx(1, 2)
qc.h(1)

qc.barrier()

qc.measure(1, 0)
qc.measure(2, 1)

fig = qc.draw('mpl')
plt.show()
```
